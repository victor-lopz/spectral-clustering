# EDO Sense Soroll

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp 
```

Algorisme de l'article de Strahan et al. 2023,
a partir de trajectòries del sistema dinàmic donat 
pel sistema d'EDOs x'=y; y'=x-x^3

```python
def edo(t, z):
    """Paràmetres:
        t: temps
        z: posició al pla R^2
    Retorna el camp vectorial del sistema d'EDOs autònom x'=y; y'=x-x^3.
    """
    x, y = z
    return np.array( [y, x - x**3] )
```

Paràmetres

```python
t_span = [0, 17]
t_steps = 1000
t_valors = np.linspace(t_span[0], t_span[1], t_steps)
condicions_inicials = [
    [-0.5, 0], [0.5, 0], [0.2, 0], [0.3, 0.3], [1, 1]
]
```

```python
def grafica_solucio(edo, t_span, t_valors, condicions_inicials, save_to_pdf=False):
    for ci in condicions_inicials:
        sol = solve_ivp(edo, t_span, ci, t_eval=t_valors)
        plt.plot(sol.y[0], sol.y[1], label=f'x(0)={ci[0]}, y(0)={ci[1]}')
    plt.xlabel('x'); plt.ylabel('y'); plt.legend(); plt.grid()
    if save_to_pdf:
        plt.savefig('edo.pdf')
    plt.show()
```

```python
grafica_solucio(edo, t_span, t_valors, condicions_inicials)
```
